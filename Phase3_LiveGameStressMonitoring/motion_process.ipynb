````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
import os
from scipy.signal import savgol_filter
````

````python
# Plot the 3 types of signals on separate axis

# Load the motion sensor data
df_acc = pd.read_csv('Data/motion_data/ACC_2024_09_20_14_54.csv')
df_gyro = pd.read_csv('Data/motion_data/GYRO_2024_09_20_14_54.csv')
df_mag = pd.read_csv('Data/motion_data/MAGN_2024_09_20_14_54.csv')

# Convert Unix timestamp (milliseconds) to datetime objects
timestamps_datetime_acc = [datetime.datetime.fromtimestamp(ts / 1000) for ts in df_acc['timestamp']]
timestamps_datetime_gyro = [datetime.datetime.fromtimestamp(ts / 1000) for ts in df_gyro['timestamp']]
timestamps_datetime_mag = [datetime.datetime.fromtimestamp(ts / 1000) for ts in df_mag['timestamp']]
df_acc['adjusted_time'] = timestamps_datetime_acc
df_gyro['adjusted_time'] = timestamps_datetime_gyro
df_mag['adjusted_time'] = timestamps_datetime_mag


# Initialize the figure and axes
fig, axs = plt.subplots(3, 1, figsize=(12, 10), sharex=True)

# Plot accelerometer data
axs[0].plot(df_acc['adjusted_time'], df_acc['x'], label='ACC X', color='blue', alpha=0.7)
axs[0].plot(df_acc['adjusted_time'], df_acc['y'], label='ACC Y', color='green', alpha=0.7)
axs[0].plot(df_acc['adjusted_time'], df_acc['z'], label='ACC Z', color='red', alpha=0.7)
axs[0].set_title("Accelerometer Signals")
axs[0].set_ylabel("Acceleration (m/s)")
axs[0].legend()
axs[0].grid()

# Plot gyroscope data
axs[1].plot(df_gyro['adjusted_time'], df_gyro['x'], label='GYRO X', color='blue', linestyle='--', alpha=0.7)
axs[1].plot(df_gyro['adjusted_time'], df_gyro['y'], label='GYRO Y', color='green', linestyle='--', alpha=0.7)
axs[1].plot(df_gyro['adjusted_time'], df_gyro['z'], label='GYRO Z', color='red', linestyle='--', alpha=0.7)
axs[1].set_title("Gyroscope Signals")
axs[1].set_ylabel("Angular Velocity (/s)")
axs[1].legend()
axs[1].grid()

# Plot magnetometer data
axs[2].plot(df_mag['adjusted_time'], df_mag['x'], label='MAGN X', color='blue', linestyle=':', alpha=0.7)
axs[2].plot(df_mag['adjusted_time'], df_mag['y'], label='MAGN Y', color='green', linestyle=':', alpha=0.7)
axs[2].plot(df_mag['adjusted_time'], df_mag['z'], label='MAGN Z', color='red', linestyle=':', alpha=0.7)
axs[2].set_title("Magnetometer Signals")
axs[2].set_ylabel("Magnetic Field (T)")
axs[2].legend()
axs[2].grid()

# Format the x-axis for all subplots
axs[2].set_xlabel("Time")
for ax in axs:
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))
    ax.tick_params(axis='x', rotation=45)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
````

````python
# Clean and preprocess all motion data


# Paths to raw data files
sensor_types = ['ACC', 'GYRO', 'MAGN']
time_measurements = ['2024_09_20_14_35', '2024_09_20_14_40', '2024_09_20_14_45', '2024_09_20_14_49', '2024_09_20_14_54']
input_dir = 'Data/Motion_data'
output_dir = 'Data/Cleaned_data'

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Function to clean and save motion sensor data
def clean_and_save_sensor_data(sensor_type, time_measurement):
    input_file = os.path.join(input_dir, f"{sensor_type}_{time_measurement}.csv")
    output_file = os.path.join(output_dir, f"{sensor_type}_{time_measurement}_cleaned.csv")

    # Load the data
    df = pd.read_csv(input_file)

    # Convert timestamps to datetime
    df['adjusted_time'] = [datetime.datetime.fromtimestamp(ts / 1000) for ts in df['timestamp']]

    # Step 1: Remove missing or invalid values
    for col in ['x', 'y', 'z']:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    df.dropna(subset=['x', 'y', 'z'], inplace=True)

    # # Step 2: Outlier detection and removal using IQR
    # for col in ['x', 'y', 'z']:
    #     Q1 = df[col].quantile(0.25)
    #     Q3 = df[col].quantile(0.75)
    #     IQR = Q3 - Q1
    #     lower_bound = Q1 - 1.5 * IQR
    #     upper_bound = Q3 + 1.5 * IQR
    #     df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

    # Step 3: Apply Savitzky-Golay filter for smoothing
    for col in ['x', 'y', 'z']:
        df[f'{col}_smoothed'] = savgol_filter(df[col], window_length=51, polyorder=3)

    # Step 4: Normalize the smoothed signals
    for col in ['x', 'y', 'z']:
        df[f'{col}_normalized'] = (df[f'{col}_smoothed'] - df[f'{col}_smoothed'].min()) / \
                                  (df[f'{col}_smoothed'].max() - df[f'{col}_smoothed'].min())

    # Save cleaned data
    cleaned_columns = ['adjusted_time', 'x_smoothed', 'y_smoothed', 'z_smoothed',
                       'x_normalized', 'y_normalized', 'z_normalized']
    df[cleaned_columns].to_csv(output_file, index=False)
    print(f"Cleaned data saved to: {output_file}")

# Iterate over all sensor types and time measurements
for sensor_type in sensor_types:
    for time_measurement in time_measurements:
        clean_and_save_sensor_data(sensor_type, time_measurement)
````
